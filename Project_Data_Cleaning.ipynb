```python
import re
```

```python
year = []
for i in MovieData_df['Year']:
    regex = re.findall('(\d+)',i)
    if regex:
        year.append(int(regex[0]))
    else:
        year.append(np.nan)
```

```python
MovieData_df['Year'] = year
```

```python
MovieData_df['Year']
```

```python
MovieData_df.Runtime = MovieData_df.Runtime.apply(lambda x: x.replace("min","")).astype('int')
```

```python
MovieData_df.Runtime
```

```python
MovieData_df.Votes = MovieData_df.Votes.apply(lambda x: x.replace(",","")).astype('int')
```

```python
MovieData_df["Gross"]
```

```python
MovieData_df["Gross"] = MovieData_df["Gross"].replace('M','',regex=True)
```

```python
MovieData_df.Gross = MovieData_df.Gross.apply(lambda x: str(x).replace("$",""))
```
