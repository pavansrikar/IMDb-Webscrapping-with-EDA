```python
#Final code for dataframing every feature into one dataframe and to run on all pages
```

```python
Title = []
Year = []
Certificate = []
Runtime = []
Genre = []
Imdb_rating = []
Metascore = []
Director = []
Votes = []
Gross = []

# for loop to run on all pages
for i in range(1,6):
    start_time = time.time()
    url = 'https://www.imdb.com/list/ls505301835/?st_dt=&mode=detail&ref_=ttls_vm_dtl&sort=alpha,asc&page={}'.format(i)
    page=requests.get(url)
    pagecontent=page.text
    soup = BeautifulSoup(pagecontent)
    
    
    for x in soup.find_all('div',attrs={'class':'lister-item-content'}):
        title = x.find('h3',attrs={'class':'lister-item-header'})
        year = x.find('span', attrs={'class': 'lister-item-year text-muted unbold'})
        certificate = x.find('span',attrs={'certificate'})
        runtime = x.find('span', attrs={'class': 'runtime'})
        genre = x.find('span',attrs={'class':'genre'})
        imdb_rating = x.find('span',attrs={'ipl-rating-star__rating'})
        metascore = x.find('span',attrs={'metascore'})
        votes = x.find('span',attrs={'name':'nv'})
            
        if title is None:
            Title.append(np.NaN)  
        else:
            Title.append(title.find('a').text)    
            
        if year is None:
            Year.append(np.NaN)
        else:
            Year.append(year.text)            
            
        if certificate is None:
            Certificate.append(np.NaN)
        else:
            Certificate.append(certificate.text)   
            
        if runtime is None:
            Runtime.append(np.NaN)
        else:
            Runtime.append(runtime.text)
            
        if genre is None:
            Genre.append(np.NaN)
        else:
            Genre.append(genre.text)
            
        if imdb_rating is None:
            Imdb_rating.append(np.NaN)
        else:
            Imdb_rating.append(imdb_rating.text)
            
        if metascore is None:
            Metascore.append(np.NaN)
        else:
            Metascore.append(metascore.text)   
            
        if votes is None:
            Votes.append(np.NaN)
        else:
            Votes.append(votes.text)
            
            
    for x in soup.find_all('div', attrs={'class': 'lister-item-content'}):
        gross = x.find_all('span', attrs={'name': 'nv'})
        if not gross:
            Gross.append(np.NaN)
        else:
            gross = gross[1].text if len(gross) > 1 else np.NaN
            Gross.append(gross)
     
    for x in soup.find_all('p',attrs={'text-muted text-small'}):
        director = x.find('a')
        if director is None:
            pass
        else:
            Director.append(director.text)
        

            
    print('page {} completed in {} seconds'.format(i,time.time()-start_time))
 
```
